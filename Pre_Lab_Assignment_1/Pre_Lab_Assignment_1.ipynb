In this pre-lab assignment we will explore the behavior in which we have two speakers emitting sound. 

The general equation government a propagating wave as a function of space and time is the following

$f(x,t) = A\sin(\frac{2\pi}{\lambda}x - \frac{2\pi}{T}t)$ 

Where $\lambda$ is the wavelength and $T$ is the period of the wave and:

$ k = \frac{2\pi}{\lambda}$ and $ Ï‰ = \frac{2\pi}{T}$

Assuming the waves are coherent and that the speakers have a relatively small offset, the amplitude of the resulting wave. 

$a =A_1\sin\left(\frac{2\pi}{\lambda} x_1 - \frac{2\pi}{T}t \right) + A_2\sin\left(\frac{2\pi}{\lambda} x_2 - \frac{2\pi}{T}t\right)$

or 

$ a = \sin(kx_1 - \omega t) + \sin(kx_2 - \omega t)$

We can do some calculations to see what this pattern looks like. First, let's calculate the wave properties of a 261 Hz wave

```
frequency = 261.                            #Hz
vs = 343.0                                  #m/s
lam = vs / frequency
k = 2 * np.pi / lam
print('Wavelength of a 261 Hz sound = ' + str(lam))
print('k of a 261 Hz sound = ' + str(k))
```

If we were to look at 2 speakers that are emitting the same frequency but with slightly different amplitudes and take a temporal snapshot **(t = 0)** of the sound waves we would see a displacement of partiples from their equilibrium poisitions that resembled a sinusoidal pattern. 

$ a_1 = A_1\sin(kx_1)$ and
$ a_2 = A_2\sin(kx_2)$

Now, lets say that we had a special device (let's say a special microphone) that we could slide across space to see that total amplitude of these 2 waves at any position at **t = 0**. What we would see is that the 2 waves interfere with one another and have positions of high "constructive" interference and low "destructive" interference. Also note that total amplitude of the total wave is the sum of the 2 individual amplitudes any particular point.

At ~0.3 m, there is a fully destructive interference, while at ~0.6 m, there fully constructive interference.

***Please note that microphones cannot pick up a negative "amplitude" so we have taken the absolute value of the sum of the amplitudes.*** 

```
A1 = 1
A2 = 0.75
x_1_positions = np.linspace(0, 4, 500)
x_2_positions = np.linspace(0, 4, 500)

amplitude_1 = A1*np.cos(k * x_1_positions)
amplitude_2 = A2*np.cos(k * x_2_positions)
amplitude_total = np.abs(amplitude_1+amplitude_2)

plt.figure(figsize = (12,8))
plt.plot(x_1_positions, amplitude_1, label = 'Wave 1')
plt.plot(x_2_positions, amplitude_2, label = 'Wave 2')
plt.plot(x_1_positions, amplitude_total, label = 'Total Amplitude')


plt.xlabel('Distance of speaker 2 (m)')
plt.ylabel('Amplitude (arbitrary)')
plt.legend()
```
