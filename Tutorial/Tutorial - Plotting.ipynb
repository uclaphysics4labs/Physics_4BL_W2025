# Two-dimensional scatter plot


We can now use our skills manipulating arrays to read in data. 

```python
import numpy as np
```

```python
# Read in data, find the right source in your google drive
read_in_array = np.loadtxt('practicelineardata.txt')
```

You can use the file explorer to examine the file and you will see a file with two columns. Remember from that if you have a two dimensional array, you can separate the colums or rows out using the : symbol, which indicates to grab all of the elements from a column.

```python
print(read_in_array)
```

```python
# The first column is the x axis, second is y axis
# Take all of the elements in 0th column
x_axis = read_in_array[:,0]
# Take all of the elements in the 1st column
y_data = read_in_array[:,1]
```

Now we will use the plotting module matplotlib.pyplot which will contains many useful functions used for plotting. As is convention, we will shorten the name of the module to plt.

```python
import matplotlib.pyplot as plt
```

We can make a scatter plot of all of the data by using the plt.scatter function, where we give the x data and the y data.

```python
# Make the scatter plot
plt.scatter(x_axis, y_data)

# Label all quantities
# Label the x axis
plt.xlabel('Some Quantity (units)')
# Label the y axis
plt.ylabel('Another Quantity (units)')
# Give the plot a title
plt.title('Test plot')
```

## Polyfit

Polyfit helps us by finding the least square polynomial fit. This means finding the best fitting curve to a given set of points by minimizing the sum of squares (the distances between the data points and the best-fit curve). It takes 3 different inputs from the user, namely x-axis data, y-axis data, and the polynomial degree.

Polyfit returns the polynomial coefficients with the coefficient of the highest power first and the other coefficients in decreasing order of the power. For example, if we try to fit a line $y=mx+b$ to a dataset, polyfit will return the slope $m$ and then the intercept $b$. 

We can store the results of polyfit in an array. The zeroth element of the array will be the best slope $m$ and the first element will be the best intercept $b$.

```python
# np.polyfit() requires at least 3 arguments or inputs
# The x-axis data, the y-axis data and the degree of the polynomial that we are trying to fit
# Let's try to fit a line to the dataset that we have read
coeff_linear=np.polyfit(x_axis,y_data,1)
# The zeroth element of coeff_linear is the slope and first element is the intercept
print(coeff_linear)
```

```python
# Since we know how to plot, we can plot our best-fit line
# coeff_lienar[0] is the slope and coeff_linear[1] is the intercept
# plt.plot() generates a connected line unlike plt.scatter() that only generates points
plt.plot(x_axis,coeff_linear[0]*x_axis+coeff_linear[1])
# Make the scatter plot of the original data
plt.scatter(x_axis, y_data)
```

We can also try to fit a second degree polynomial or a quadratic curve to this data and see what that looks like. We follow the same procedure as before but with a second degree polynomial $ax^2+bx+c$

```python
# Let's try to fit a quadratic curve to the dataset that we have read
coeff_quad=np.polyfit(x_axis,y_data,2)
# The zeroth element of coeff_quad is "a", first element is "b" and the second element is "c"
print(coeff_quad)
```

```python
# Since we know how to plot, we can also plot our best-fit quadratic curve
# coeff_quad[0] is "a", coeff_quad[1] is "b" and coeff_quad[2] is "c"
# plt.plot() generates a connected line unlike plt.scatter() that only generates points
plt.plot(x_axis,coeff_quad[0]*x_axis**2+coeff_quad[1]*x_axis+coeff_quad[2])
# Make the scatter plot of the original data
plt.scatter(x_axis, y_data)
```

```python
plt.scatter(x_axis, y_data,color="green",label="Data")
plt.plot(x_axis,coeff_quad[0]*x_axis**2+coeff_quad[1]*x_axis+coeff_quad[2],color="red",label="Fit")
plt.legend()
```

We see that it still looks similar to a line because "a" is -3.59166169e-03

# Plot customizations

Run the following plots to see different customizations we can do

```python
# Generate an array from 0 to 20 with 100 points
x_axis = np.linspace(0., 20., 100)
# Generate a sample data
y_axis = 8. * x_axis + 5. 

# plt.plot plots a smooth line in between the data
plt.plot(x_axis, y_axis)
plt.scatter(x_axis, y_axis)
plt.title("Silly title")
plt.xlabel("Example line (units)")
plt.ylabel("Another example (other units)")
```

[Here is the list of all of the colors you can use](https://matplotlib.org/3.1.0/gallery/color/named_colors.html)

```python
# Generate a second sample
y_axis_2 = x_axis**2. + 1.

# Plot both at once, change the colors
plt.plot(x_axis, y_axis, color = 'purple')
plt.plot(x_axis, y_axis_2, color = 'green')
```

```python
coeff_linear=np.polyfit(x_axis,y_axis,1)
print(coeff_linear)
```

```python
coeff_quad=np.polyfit(x_axis,y_axis_2,2)
print(coeff_quad)
```

```python
plt.plot(x_axis,coeff_linear[0]*x_axis+coeff_linear[1])
```

Whenever you have more one line or data plotted, always include a legend

```python
# Plot both at once, change the colors
plt.plot(x_axis, y_axis, color = 'purple', label = "Eggs")
plt.plot(x_axis, y_axis_2, color = 'green', label = "Spam")

# Create a legend
plt.legend()
```

```python
# Generate more datasets 
x_axis = np.linspace(0., 30, 100)
y_axis = x_axis + 10
y_axis_2 = x_axis + 20
y_axis_3 = x_axis + 30

plt.plot(x_axis, y_axis, color = 'purple', linestyle = 'dashed', label = "Eggs")
plt.plot(x_axis, y_axis_2, color = 'green', linestyle = 'dotted', label = "Spam")
plt.plot(x_axis, y_axis_3, color = 'orange', linewidth = 3.0, label = "More spam")
plt.legend()
```
